```python
import requests
import base64
from collections import defaultdict
from pathlib import Path
import pickle

endpoint = 'https://api-us.faceplusplus.com'

#先頭がorgが始まるディレクトリを対象とする
face_dict = defaultdict(str)
src = sorted(Path('/xxx').glob('*.jpg'))

for i, file in enumerate(src):
    if i <= 4000:
        continue
    print("{}:{}".format(i,file))
    img_file = base64.encodestring(open(file, 'rb').read())
    response = requests.post(
        endpoint + '/facepp/v3/detect',
        {
            'Content-Type': 'multipart/form-data',
            'api_key': "xxx",
            'api_secret': "xxx",
            'image_base64': img_file,
            'return_attributes': 'gender,age,headpose,facequality,eyestatus,emotion,ethnicity,beauty,mouthstatus'
        }
    )
    try:
        result = response.json()
        #写真に顔が1つしか写っていないものを対象とする
    
        if len(result["faces"]) == 1:
            face_dict[file.name[:-4]] = result
    except:
        print("{}:{}でエラーが発生".format(i,file))
    
    if i % 1000 == 0 and i != 0:
        with open('/xxx', 'wb') as f:
            pickle.dump(face_dict, f)
```

```python
import mysql.connector
import pandas as pd
import pandas.io.sql as psql
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)
import numpy as np
import pickle
import shutil
import re
from pathlib import Path
from collections import defaultdict
```

```python
target_pictures = {}
src = sorted(Path('xxx').glob('**/*.dump'))
for i in src:
    with open(i, 'rb') as f:
        tmp = pickle.load(f)
        target_pictures.update(tmp)
target_pictures = sorted(target_pictures.items(), key=lambda x:x[0])
```

```python
#顔が写っていない写真を抽出
src = sorted(Path('xxx').glob('*.jpg'))
for i in src:
    no_faces = [str(x.name[:-6]) for x in i.iterdir()]
    for i in target_pictures:
        if i[0][:-2] in no_faces:
            del target_pictures[i]
```

```python
for i, v in enumerate(target_pictures):
    if target_pictures[i][1]["faces"][0]["attributes"]["mouthstatus"]["surgical_mask_or_respirator"] >= 90:
        del target_pictures[i]
        continue
    if target_pictures[i][1]["faces"][0]["attributes"]["glass"]["value"] == "Dark":
        del target_pictures[i]
        continue
    if target_pictures[i][1]["faces"][0]["attributes"]["eyestatus"]["left_eye_status"]["occlusion"] >= 90 or target_pictures[i][1]["faces"][0]["attributes"]["eyestatus"]["right_eye_status"]["occlusion"] >= 90:
        del target_pictures[i]
        continue
    if target_pictures[i][1]["faces"][0]["attributes"]["eyestatus"]["left_eye_status"]["no_glass_eye_close"] >= 90 or target_pictures[i][1]["faces"][0]["attributes"]["eyestatus"]["right_eye_status"]["no_glass_eye_close"] >= 90:
        del target_pictures[i]
        continue
    if target_pictures[i][1]["faces"][0]["attributes"]["eyestatus"]["left_eye_status"]["normal_glass_eye_close"] >= 90 or target_pictures[i][1]["faces"][0]["attributes"]["eyestatus"]["right_eye_status"]["normal_glass_eye_close"] >= 90:
        del target_pictures[i]
        continue
    if target_pictures[i][1]['faces'][0]['face_rectangle']['top'] <= -55:
        del target_pictures[i]
        continue
```

```python
pre_id = ""
score_dict = defaultdict(list)
work_dict = {}
for v in target_pictures:
    before_under_bar = v[0].find("_")
    id = v[0][:before_under_bar]
    if id != pre_id and work_dict != {}:
        #idが変わったら、値が一番大きいkeyとvalueを設定
        max_kv = max(work_dict.items(), key=lambda x: x[1])
        score_dict[max_kv[0]] = round(max_kv[1], 2)
        work_dict.clear()
    #idが変わるまで辞書に入れる
    work_dict[v[0]] = (v[1]['faces'][0]['attributes']['beauty']['female_score'] + v[1]['faces'][0]['attributes']['beauty']['male_score']) / 2
    pre_id = id
```

```python
#一番スコアの高いファイルを集める
#初回のみ使用
for i in score_dict.keys():
    target = sorted(Path('xxx').glob('**/*' + str(i) + '*'))
    shutil.copy(str(target[0]), "xxx")
```

```python
#目視で確認するデータを対象のフォルダへ移動
no_faces_new = [x[:-2] for x in no_faces]
for i, n in score_dict.items():
    if i[:-4] in no_faces_new:
        file_paths = Path('/xxx').glob('**/' + i + '.jpg')
        for file_path in file_paths:
            shutil.copy(str(file_path), "/xxx")
```

```python
id_score = {}
df_list = []
for i, v in score_dict.items():
    key = re.match(r'(.*)_(.*)', i)
    data_list = [int(key.group(1)), v]
    df_list.append(data_list)
```

```python
id_score = pd.DataFrame(list(df_list))
```

```python
class Pairs:
    def __init__(self):
        self._cnx = mysql.connector.connect(user='xxx', password='xxx', database='pairs')
        self._cursor = self._cnx.cursor()
```

```python
p = Pairs()
```

```python
user = psql.read_sql("select id, good from users;", p._cnx)
user.replace('500+', 500, inplace=True)
user = user.astype({'good': 'int16'})
```

```python
user = user.set_index("id")
face_score = id_score.rename(columns={0: 'id', 1: 'score'})
face_score = face_score.set_index("id")
data = pd.concat([user, face_score], axis=1, join="inner")
sns.distplot(data["score"], kde=False, rug=False, bins=10)
```

```python
data["good_range"] = "" 
for index, row in data.iterrows():
    if 0 <= row["good"] <= 49:
        data.at[index, 'good_range'] = "0 ~ 49"
    elif 50 <= row["good"] <= 99:
        data.at[index, 'good_range'] = "50 ~ 99"
    elif 100 <= row["good"] <= 149:
        data.at[index, 'good_range'] = "100 ~ 149"
    elif 150 <= row["good"] <= 199:
        data.at[index, 'good_range'] = "150 ~ 199"
    elif 200 <= row["good"] <= 249:
        data.at[index, 'good_range'] = "200 ~ 249"
    elif 250 <= row["good"] <= 299:
        data.at[index, 'good_range'] = "250 ~ 299"
    elif 300 <= row["good"] <= 349:
        data.at[index, 'good_range'] = "300 ~ 349"
    elif 350 <= row["good"] <= 399:
        data.at[index, 'good_range'] = "350 ~ 399"
    elif 400 <= row["good"] <= 449:
        data.at[index, 'good_range'] = "400 ~ 449"
    elif 450 <= row["good"] <= 499:
        data.at[index, 'good_range'] = "450 ~ 499"
    else:
        data.at[index, 'good_range'] = "500 ~"
```
